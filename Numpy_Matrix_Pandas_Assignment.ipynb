## Q1. Create a 3×3 NumPy Matrix

### Input
```
1 2 3
4 5 6
7 8 9
```

### Output
```python
[[1 2 3]
 [4 5 6]
 [7 8 9]]
```


````python
import pandas as pd
import numpy as np

matrix = np.arange(1,10).reshape(3,3)
print(matrix)
````
