## Q1. Sort Age in Ascending Order

Input:
Age = [25, 18, 30, 22]

Output:
18
22
25
30

```python
import pandas as pd

df = pd.DataFrame({
    'Age':[25,18,30,22]
})
df
df.sort_values(by='Age')
```

## Q15. Mini Practice

Create a DataFrame with:
Name, Age, Score, Height

Tasks:
1. Sort by Score (Descending)
2. Find Mean Age
3. Find Median Height
4. Show Score Frequency

```python
df = pd.DataFrame({
    'Name':['Sandeep','Suman','Riju','Vishesh'],
    'Age':[50,60,30,40],
    'Score':[70,65,90,70],
    'Height':[5.9,5.2,5.5,6.5]
})
df
print(df.sort_values(by='Score', ascending=False))
print(df.Age.mean(numeric_only=True))
print(df.Height.median(numeric_only=True))
print(df.Score.value_counts())

```
