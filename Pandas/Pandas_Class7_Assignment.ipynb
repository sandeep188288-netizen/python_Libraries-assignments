# Q1. Create a Student DataFrame

## Problem Statement

# Q8. Remove Duplicate Rows

## Problem Statement
Create the following DataFrame:

| Name | Age | City | Marks |
|------|----:|------|------:|
| Aman | 21 | Bhopal | 85 |
| Rahul | 22 | Indore | 72 |
| Aman | 21 | Bhopal | 85 |
| Neha | 20 | Delhi | 91 |

Remove duplicate rows using **drop_duplicates()**.

## Sample Input
```python
Use the table above.
```

## Sample Output
```python
Duplicate rows removed.
```

````python
df = pd.DataFrame({
    'Name':['Aman','Rahul','Aman','Neha'],
    'Age':[21,22,21,20],
    'City':['Bhopal','Indore','Bhopal','Delhi'],
    'Marks':[85,72,85,91]
})
df
df.drop_duplicates(inplace=True)
df
````

# Q9. Fill Missing Values

## Problem Statement
Create the following DataFrame:

| Name | Marks |
|------|------:|
| Aman | 85 |
| Rahul | NaN |
| Neha | 91 |
| Priya | NaN |

Fill missing values in the **Marks** column with **0** using **fillna()**.

## Sample Input
```python
Use the table above.
```

## Sample Output
```python
Missing values replaced with 0.
```

````python
df = pd.DataFrame({
    'Name':['Aman','Rahul','Neha','Priya'],
    'Marks':[85,None,91,None]
})
df
# df.isnull().sum()       #count null column wise
df['Marks'].fillna(0, inplace=True)
print(df)
````

# Q10. Final Practice

## Problem Statement
Create a DataFrame with the following data:

| Name | City | Marks |
|------|------|------:|
| Aman | Bhopal | 85 |
| Rahul | Indore | 72 |
| Neha | Delhi | 91 |
| Priya | Bhopal | 68 |
| Ravi | Indore | 95 |

Perform all of the following:
1. Display students with Marks > 80.
2. Sort the DataFrame by Marks.
3. Find the average Marks of each City.

## Sample Input
```python
Use the table above.
```

## Sample Output
```python
Required outputs displayed.
```

````python
df = pd.DataFrame({
    'Name':['Aman','Rahul','Neha','Priya','Ravi'],
    'City':['Bhopal','Indore','Delhi','Bhopal','Indore'],
    'Marks':[85,72,91,68,95]
})
df

print(df[df.Marks > 80])
print(df.sort_values(by='Marks'))
print(df.groupby('City').Marks.mean())
````
