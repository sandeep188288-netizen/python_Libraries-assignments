# Q1. Select Single Column

## Problem Statement
Create a DataFrame and display only the Name column.

## Sample Input
```python
Name, Age
```

## Sample Output
```python
Only Name column
```


````python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Rahul','Riya','Priya','Karan','Arjun'],
    'Age':[20,23,22,27,24]
})
df
print(df.Name)

````


# Q2. Select Multiple Columns

Display students belonging to Bhopal.

Matching rows

Count occurrences of each city.

City column

Frequency count


# Q9. Group By City

## Problem Statement
Find average marks city-wise.

## Sample Input
```python
City and Marks
```

## Sample Output
```python
Grouped result
```


````python
df = pd.DataFrame({
    'Name': ['Rahul','Riya','Priya','Karan','Arjun'],
    'Age':[20,19,22,18,24],
    'Marks': [90,95,80,50,70],
    'City':['Bhopal','Maihar','Bhopal','Rewa','Katni']
})
df
# df['Marks'].mean()
print(df.groupby('City')['Marks'].mean())
````
