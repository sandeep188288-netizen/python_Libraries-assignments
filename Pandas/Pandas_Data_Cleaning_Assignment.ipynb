## Q1. Find Missing Values

Input:
df = pd.DataFrame({
    "Name":["Aman","Riya",None],
    "Age":[21,np.nan,25]
})

Output:
    Name    Age
0  False  False
1  False   True
2   True  False

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["Aman", "Riya", None],
    "Age": [21, np.nan, 25]
})
df

print(df.isna())
```

## Q3. Fill Missing Age with 0

Input:
df = pd.DataFrame({

## Q13. Mini Data Cleaning Task

Input:
data = {
    "Name":["Aman","Raj","Aman",None],
    "Age":[20,None,20,25],
    "City":["Bhopal","Delhi","Bhopal","Indore"]
}

Task:
1. Fill missing Age with mean.
2. Remove duplicate rows.
3. Display final DataFrame.

```python
data = pd.DataFrame({
    "Name": ["Aman", "Raj", "Aman", None],
    "Age": [20, None, 20, 25],
    "City": ["Bhopal", "Delhi", "Bhopal", "Indore"]
})
data

data.Age.fillna(data.Age.mean(), inplace = True)
data.drop_duplicates(inplace = True)
print(data)
```
