```python
Use NumPy arrays
```

## Problem Statement

## Sample Input
```python
Use NumPy arrays
```

## Sample Output
```python
Expected Output
```


````python
a = np.array([2,4,43,15,32,9,3,11,60])
a
print(a[a > 15])            #give value
print(np.where(a > 15))     #give index
````
